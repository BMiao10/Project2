**1. Explain the distance metric you utilized to calculate the similarity/dissimilarity between small molecules.** 

I used bitwise Jaccard distance (1 - Jaccard index) to calcuate scores in hierarchical clustering, since all values are binary, and this is one of the fastest distance metrics to use. The Jaccard distance measures 1 - (intersection / union), so if there are many matching onbits, the intersection is high. This means the Jaccard index is high and the distance is low (closer to 0), which makes sense because these values would be considered to be more similar. If there is no intersection between any values, the intersection and Jaccard index is 0, and the distance is at its max of 1. In cases where both values are 0, my implementation returns a distance of 0 because they are technically the same, although the Jaccard index would be undefined.

For partitioning clustering, I use euclidean distance which is unfortunately slower than Jaccard index, but is necessary to accomodate for non-0 and 1 values in the centroids I create. Euclidean distance measures the "shortest path" distance between points (= sqrt of sum of squared pairwise differences between points), and values can range from 0 to ~inf depending on the range of the values within each point. Points that are identical have a difference of 0 so a euclidean distance of 0, while points that are far apart have large pairwise differences and thus large euclidean distances.

```python
### Partition Clustering

# read in the ligands
ligands = algs.readLigands("./ligand_information.csv")

# create partitioning clustering object
hier = algs.PartitionClustering()

# run partioning clustering with multiple values of k and store results
all_results = np.empty((0,len(ligands)))

for k in [2,4,8,15,20,50,100]:
    hier.k = k
    results = hier.cluster(ligands, dist_met="euc", preprocess=algs.preprocessLigands)

    all_results = np.vstack((all_results, np.array(results)))

# save results to file
np.save("./output/partition_results.npy", all_results)
```

**4. Explain your choice of partitioning clustering algorithm. Is it sensitive to initialization conditions? How do you select the number of clusters?**  
The partitioning algorithm I chose to implement was kmeans, with centroid initalization by kmeans++. kmeans is considered to be a relatively fast clustering algorithm, with an estimated runtime of O(k\*n) where n is the number of samples to be clustered and k is the number of output clusters.  

kmeans does depend on how quickly clusters converge, which can be challenging if bad centroids are initialized. One issue is that if centroids are too close together, they can cause certain samples to bounce between their repective clusters and prevent the algorithm from converging. To combat this issue, I added in a term for the maximum number of iterations the algorithm should run through if it does not converge before then. Additionally, I used kmeans++ to assign my initial centroids rather than randomly since kmeans++ chooses centroids that are far apart, and tends to converge faster. 

When I update my centroids, I create new centroids by taking the pairwise mean of the points in that centroid's cluster. This means that some centroids are non-binary arrays, so I use euclidean distance, rather than a bitwise Jaccard index to calculate the distance from each point to the centroids as I update the clusters. I then check whether any clusters have no points, and assign the first available point from another cluster to it, which ensures that no clusters are left empty at the end. 

Looking at the silhouette scores for various k, the quality of kmeans clustering does appear to be sensitive to initialization for k. I chose values of [2,4,8,15,20,50,100] to test and had silhouette scores ranging from 0.1 to 0.38. The higher the silhouette score, the better the clustering. A high silhouette score indicates low cohesion, or distance between points in a cluster, and high separation, or distance from a point to other clusters. k=100 had the highest silhouette score of 0.37 and it is possible that testing more values of k might lead to ever higher silhouette scores, but given that the ligands being tested here mostly clustered into only a few large clusters by UMAP, having a lower k might be more reasonable. 

Since k=4 gave a local maximum score of 0.338, I used it as the best k for partitioning clustering related questions.


```python
### Partition clustering silhouette scores

# Read in the ligands
ligands = algs.readLigands("./ligand_information.csv")

# Get the partition clustering results
results = np.load("./output/partition_results.npy")

# Initialize params
output_file = "./output/full_dist_mat.npy"

# Create empty list to store silhouette scores
scores = []

# Extract out first results (k=2, as described above) and find the silhouette score
# This will also create a distance matrix for the full set of ligands and only needs to be run once 
curr_result = results[0]
curr_score = algs.calcClusterQuality(samples=ligands, clust_labels=curr_result, preprocess=algs.preprocessLigands, output_matrix=output_file)
scores.append(curr_score)

# Get the silhouette scores for all other results, using the precomputed distance matrix 
input_file = "./output/full_dist_mat.npy"

for curr_result in results[1:]:
    curr_score = algs.calcClusterQuality(samples=ligands, clust_labels=curr_result, preprocess=algs.preprocessLigands, input_matrix=input_file)
    print(curr_score)
    scores.append(curr_score)
    
# map scores to k values and save to csv
scores_df = pd.DataFrame([[2,4,8,15,20,50,100],scores]).T
scores_df.columns = ["k", "silhouette"]
scores_df.to_csv("./output/Q4-partition_silhouette_scores.csv")
```

**5. Cluster the small molecules using your implementation of a hierarchical clustering algorithm. Visualize this clustering.**  

Hierarchical clustering is performed below with a random subsample of 1000 ligands (seed set to ensure reproducibility). I used complete linkage for this clustering, and Jaccard Index as my distance metric. 

The UMAP was colored with k=8 clusters. k = 8 was chosen because it had the largest silhouette coefficient for any k tested under 15, which would have been too many colors to visualize. Additionally, the UMAP showed that the subset only had a few large clusters, so a lower k already captures a large amount of variance.

Clustering results are saved to the output folder as a numpy file called "hier_1000subset_results.npy"  
The UMAP representation of the clusters was saved to output folder as "Q5-UMAP_hierarchical.png"  



```python
### Hierarchical clustering

# read in the ligands
ligands = algs.readLigands("./ligand_information.csv")

# subset the ligands (selection without replacement) to use for hierarchical clustering
random.seed(3)
subset_ind = random.sample(range(len(ligands)), 1000)
ligands_subs = [ligands[i] for i in subset_ind]

# create hierarchical clustering object
hier = algs.HierarchicalClustering()

# initialize empty array to contain clustering results
all_results = np.empty((0,len(ligands_subs)))

# cluster for various values of k
for k in [2,4,8,15,20,50,100, 150,200]:
    hier.k = k
    results = hier.cluster(ligands_subs, preprocess=algs.preprocessLigands, linkage = "min", dist_met="bit") 
    all_results = np.vstack((all_results, np.array(results)))

# save results to numpy file
np.save("./output/hier_1000subset_results.npy", all_results)
```

```python
### Plotting UMAP for hierarchical clustering

# read in the ligands
ligands = algs.readLigands("./ligand_information.csv")

# subset the ligands (selection without replacement) to use for hierarchical clustering
random.seed(3)
subset_ind = random.sample(range(len(ligands)), 1000)
ligands_subs = [ligands[i] for i in subset_ind]

# extract values to np.array
onbits = algs.preprocessLigands(ligands_subs)

# create UMAP object and apply transformation to sample values
reducer = umap.UMAP(n_neighbors=10)
embedding = reducer.fit_transform(onbits)

# store umap data in dataframe
embed_df = pd.DataFrame(embedding)
embed_df.columns = ["UMAP1", "UMAP2"]

# Get the partition clustering results for k = 8
results = np.load("./output/hier_1000subset_results.npy")
embed_df["cluster"] = results[2]
embed_df = embed_df.sort_values("cluster")
embed_df["cluster"] = embed_df["cluster"].astype(int).astype(str)

# plot
fig, ax = plt.subplots(figsize=(10,10))
sns_plot = sns.scatterplot(data = embed_df, x="UMAP1", y="UMAP2", hue="cluster", legend=True)
plt.legend(bbox_to_anchor=(1., 1), loc=2, borderaxespad=0.)

# save figure
fig = sns_plot.get_figure()
fig.savefig("./output/Q5-UMAP_hierarchical.png") 
```

**6. Explain your choice of hierarchical clustering algorithm. Is it sensitive to initialization conditions? How do you select the number of clusters?**

My hierarchical clustering was performed by assigning each point to its own cluster and joining clusters together (agglomerative clustering) until a certain number of clusters remained. The clustering was performed with complete linkage, which is expected to work better with "globular" data**\*\*** and the UMAP showed that this data could be projected into mostly globular structures. Complete linkage updates cluster values to be the maximum distance between points in the cluster and other points. Thus, when a point *P* is chosen as having the minimum distance in the matrix, it is certain that this point is within a certain distance to all points in that cluster.

**\*\*** https://scikit-learn.org/stable/auto_examples/cluster/plot_linkage_comparison.html  


The number of clusters can be selected based on the silhouette score (higher is better). Out of the values of k that I tested for, 200 had the highest silhouette score. In theory, this would be one way to choose k for the clustering.

However, initialization, including the choice of k in this particular implementation and subset of ligands does not seem to affect silhouette score too much, since the values of k tested ( [2,4,8,15,20,50,100,150,200] ) only resulted in a range of silhouette scores from 0.32 - 0.48. 

k may be better determined through domain knowledge of how these structures should cluster or other methods like the elbow method, rather than the silhouette score in this case. For example, for the UMAP, which is focused on visualization, I chose a lower value of k that did not have the optimal silhouette score. This was also guided by the fact that having 200 clusters did not seem to be particularly useful when the subset was only 1000 ligands.

Since it looked better in the UMAP visualization, gave a reasonable silhouette score, and made more sense in the context of trying to cluster the ligand into manageable categories, I used k=8 as the "best" clustering for my hierarchical clusters.



```python
### Calculating silhouette scores for hierarchical clustering

# read in the ligands
ligands = algs.readLigands("./ligand_information.csv")

# subset the ligands (selection without replacement) to use for hierarchical clustering
random.seed(3)
subset_ind = random.sample(range(len(ligands)), 1000)
ligands_subs = [ligands[i] for i in subset_ind]

# extract values to np.array
onbits = algs.preprocessLigands(ligands_subs)

# Get the partition clustering results
results = np.load("./output/hier_1000subset_results.npy")

# Segment the distance matrix for just the subset of ligands
full_mat = np.load("./output/full_dist_mat.npy")
full_mat = full_mat[np.array(subset_ind)[:, None], np.array(subset_ind)]

# Save subsetted distance-matrix to numpy file
np.save("./output/subset_dist_mat.npy", full_mat)

# Create empty list to store silhouette scores
scores = []

# Get the silhouette scores for all other results, using the precomputed distance matrix 
input_file = "./output/subset_dist_mat.npy"

for curr_result in results:
    curr_score = algs.calcClusterQuality(samples=ligands_subs, clust_labels=curr_result, preprocess=algs.preprocessLigands, input_matrix=input_file)
    scores.append(curr_score)

# map scores to k values and save to csv
scores_df = pd.DataFrame([[2,4,8,15,20,50,100,150,200],scores]).T
scores_df.columns = ["k", "silhouette"]
scores_df.to_csv("./output/Q6-hierarchical_silhouette_scores.csv") 

scores_df

```

**7. Evaluate the quality of both clusterings using your implementation of a clustering quality metric. Explain your choice of quality metric. Which clustering performed ‘best’ according to your metric?**

I used silhouette score to measure the quality of the clusters (see Q4/6 for use of implementation). To calculate the silhouette score for each point *i*, I calculate **a** = the average distance from *i* to all the other members of its cluster and **b** = the average distance from *i* to all the other members of the closest cluster. 

**a** decreases with better clustering because it indicates that all members of the cluster are closer together. **b** increases with better clustering because it means that even the closest cluster is far away. 

Silhouette score = (b - a) / max(b,a), so as **b** increases and **a** decreases, the silhouette score increases and indicates better clustering quality. 

Silhouette scores in theory should not be compared directly between clustering algorithms, since algorithms optimize for different objectives (eg. shortest euclidean vs jaccard distance) that may skew silhouette scores. Different algorithms may also have an optimal silhouette score at different values of k. Also in this case, the hierarchical clustering was run on a subset of 1000 ligands, while partitioning was run on the whole set of ~9000 ligands.

If we were to compare, such as in the cell below, we can look at silhouette scores between different values of k for Hierarchical vs Partition clustering. On average, hierarchical clustering has better silhouette scores and also has the highest silhouette score of 0.47 at k = 200, which I suppose could be argued is the "best" clustering if only considering this metric. 

For the sake of not trying to measure 200 different clusters, I will stick with my original "best" clustering of hierarchical clustering with k=8.


```python
# read in clustering silhouette scores
part = pd.read_csv("./output/Q4-partition_silhouette_scores.csv", index_col=0)
hier = pd.read_csv("./output/Q6-hierarchical_silhouette_scores.csv", index_col=0)

# rename columns so they don't have duplicate names
part.columns = ["k", "part"]
hier.columns = ["k", "hier"]

# merge data frames and plot
plot_df = part.merge(hier, how="outer", left_on="k", right_on="k")

# print / display values
print("Hierarchical average score = %f"%np.mean(plot_df["hier"]))
print("Partitioning average score = %f"%np.mean(plot_df["part"]))
plot_df

```

**8. Compare the two clusterings using your implementation of clustering similarity. How similar are the two clusterings using this function?**

Cluster similarity can be calculated using the Jaccard Index - this is a measure of the intersection of two groups over their union and is a ratio of how many objects in the first group match the second. For cluster results specifically, the objects are pairs of points. If points A and B are both located in the same cluster in both results, then they are considered a "true positive" / match, and is counted as part of the intersection. The union is all pairs that occur in both sets of results, except for pairs that are found in different clusters in both sets. These are "true negatives" and not counted as part of the union for Jaccard (although it is for Rand).

Because the Jaccard Index considers the ratio of "true positives" to total pairs between different clustering algorithms, it can give some indication of whether the results are accurate or not. A high Jaccard Index would be close to 1 - with an index of 1 indicating all the pairs are clustered together in both sets of results - and low Jaccard Index is close to 0. 

The "best" clustering results from hierarchical and partioning clustering were when these algorithms were initialized with k=8 and k=4, respectively, as discussed in previous questions. Only the subset of 1000 ligands were considered, since there would be no overlap between ligands not run in hierarchical clustering, and the Jaccard index was 0.440. This is a mediocre Jaccard Index, indicating about 40% of pairs "correctly" clustered. However, given that these clusterings were from different algorithms with different distance metrics and the partition clustering was run on the full set of ligands, it is not surprising that the Jaccard Index is on the lower side.

I also looked at clustering between different k's of hierarchical clustering, which should show more similarity. Comparing k=4 and k=8, the Jaccard index was 0.52, not too much higher than with partitioning clustering. k=8 was much more similar to k=20, with Jaccard index = 0.71, which also suggests that k initialization may be less important in hierarchical clustering, since the silhouette scores between these clusterings was also similar.


```python
# get the subset of ligands since only the subset is run for hierarchical
# read in the ligands
ligands = algs.readLigands("./ligand_information.csv")
random.seed(3)
subset_ind = random.sample(range(len(ligands)), 1000)

# load the results from hierarchical and partitioning clustering
part = np.load("./output/partition_results.npy")
hier = np.load("./output/hier_1000subset_results.npy")

# get the "optimal" clustering for both 
# use k = 4 for partitioning and k = 8 for hierarchical
part_4 = part[1]
hier_8 = hier[2]

# subset the partitioning clustering
part_4 = part_4[subset_ind]

# compare the similarity of clustering using the Jaccard Index
print(algs.calcClusterSimilarity(part_4, hier_8))

# also compare hier_8 to other k initializations of hier for fun
hier_4 = hier[1]
hier_20 = hier[4]

print(algs.calcClusterSimilarity(hier_4, hier_8))
print(algs.calcClusterSimilarity(hier_20, hier_8))

```
